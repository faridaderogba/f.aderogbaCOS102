```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load dataset
data = pd.read_csv("GIG-logistics.csv")

# Function to check employee
def check_employee():
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    # Filter dataset
    result = data[
        (data['Name'].str.lower() == name) &
        (data['Department'].str.lower() == dept)
    ]

    if not result.empty:
        # Employee exists
        messagebox.showinfo("Success", f"Welcome {name.title()}!")

        # Show others in same department
        others = data[
            (data['Department'].str.lower() == dept) &
            (data['Name'].str.lower() != name)
        ]

        if not others.empty:
            members = "\n".join(others['Name'])
            messagebox.showinfo("Department Members", members)
        else:
            messagebox.showinfo("Department Members", "No other members found.")

    else:
        messagebox.showerror("Error", "Employee does not exist.")

# GUI setup
root = tk.Tk()
root.title("GIG Logistics Verification")

tk.Label(root, text="Enter Name").pack()
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department").pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Check Employee", command=check_employee).pack()

root.mainloop()
```
