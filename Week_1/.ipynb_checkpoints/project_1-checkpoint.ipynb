```python
# Simple interest
print("Simple Interest Calculator")

P = float(input("Enter the principal amount: "))
R = float(input("Enter the rate of interest (%): "))
T = float(input("Enter the time (years): "))

A = P * (1 + (R/100) * T)

print("The total amount is:", A)
```

```python
# Compound Interest 
print("Compound Interest Calculator")

P = float(input("Enter the principal amount: "))
R = float(input("Enter the interest rate: "))
n = int(input("Enter number of times interest is applied per year: "))
t = float(input("Enter time in years: "))

A = P * (1 + (R/n)) ** (n*t)

print("The total amount is:", A)
```

```python
# Annuity plan
print("Annuity Plan Calculator")

PMT = float(input("Enter payment amount each period: "))
R = float(input("Enter interest rate: "))
n = int(input("Enter number of payments per year: "))
t = float(input("Enter number of years: "))

A = PMT * ((1 + R/n) ** (n*t) - 1) / (R/n)

print("The future value of the annuity is:", A)
```
