```python
PSEUDOCODE:

START

SCAN items
CALCULATE total + tax
DISPLAY total

ASK for payment method

IF payment method = CASH THEN
    COUNT cash given

    IF cash >= total THEN
        PRINT receipt
    ELSE
        REQUEST more cash
    ENDIF

ELSE IF payment method = CARD THEN
    PROCESS card payment

    IF transaction successful THEN
        PRINT receipt
    ELSE
        RETRY payment
    ENDIF

ENDIF

END
```
